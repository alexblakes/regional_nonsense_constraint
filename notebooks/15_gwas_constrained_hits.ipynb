The problem is that in the GWAS catalogue, variants consequences are annotated by the most severe consequence across any transcript. In the most biologically relevant transcript they are much more likely to be missense or synonymous. 

230) Is the only reported frameshift variant in the list. It may represent 12-120994314-GC-G, which is a known variant in HNF1A pathogenic for monogenic diabetes. A colocated synonymous variant (12-120994314-G-C) is a significant GWAS hit for type 2 diabetes, but is not pathogenic for monogenic diabetes.

The scarcity of clearcut frameshift variants is consistent with variant mis-annotation.

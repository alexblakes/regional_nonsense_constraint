# Constraint plots

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.lines import Line2D
from scipy.stats import norm

# Set style and color
%run _color_tools.ipynb
```

## Load data

```python
z_loeuf = pd.read_csv("../outputs/expected_variants_all_regions_stats.tsv", sep="\t")
```

## Plotting function

```python
def plot_z_loeuf(df, ax):
    m1 = df["csq"] == "nonsense"
    m2 = df["region"] == "transcript"
    m3 = df["z"].between(-5, 1)

    data = df[m1 & m2 & m3]
    x = data["z"]
    y = data["loeuf"]

    ax.hexbin(x, y, cmap="Blues", mincnt=1, gridsize=40)

    ax.set_xlabel("Nonsense Z score"),
    ax.set_ylabel("LOEUF")

    # ax.set_xlim(-6, 2)

    ax.text(
        x=0.95,
        y=0.05,
        s=r"rho = 0.77",
        ha="right",
        va="bottom",
        transform=ax.transAxes,
    )
```

## Plot figure

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 3))
plot_z_loeuf(z_loeuf, ax)
```

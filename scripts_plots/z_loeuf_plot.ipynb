```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.lines import Line2D
from scipy.stats import norm

# Set style and color
%run _color_tools.ipynb
```

## Load data

```python
z_loeuf = pd.read_csv("../outputs/expected_variants_all_regions_stats.tsv", sep="\t")
```

```python
def plot_z_loeuf(df, ax):
    m1 = df["csq"] == "nonsense"
    m2 = df["region"] == "transcript"
    m3 = df["z"].between(-5, 2)

    data = df[m1 & m2 & m3]
    x = data["z"]
    y = data["loeuf"]

    ax.hexbin(x, y, cmap="Blues", mincnt=1)
    
    ax.set_xlabel("Nonsense Z score"),
    ax.set_ylabel("LOEUF")
    
    #ax.set_xlim(-6, 2)
    
    ax.text(x=0.6, y=0.22, s=r"$\rho$ = 0.77", fontsize="medium")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3.5, 3))
plot_z_loeuf(z_loeuf, ax)
```

```python
def plot_z_loeuf(df):
    m1 = df["csq"] == "nonsense"
    m2 = df["region"] == "transcript"
    m3 = df["z"].between(-5, 2)

    data = df[m1 & m2 & m3]
    x = data["z"]
    y = data["loeuf"]

    g = sns.jointplot(x=x, y=y, kind="hex")
    g.set_axis_labels(x="Nonsense Z score", y = "LOEUF")
    
    #ax.set_xlim(-6, 2)
    
    g.text(x=0.6, y=0.22, s=r"$\rho$ = 0.77", fontsize="medium")

    return g
```
